```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/IMDB Dataset.csv', sep = ',')
Review = df[['review']].head(1000)
Review
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = Review['review']
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
tfidf = TfidfVectorizer(smooth_idf=True,norm='l1')
print(tfidf)
```

```python
from keras.preprocessing.text import text_to_word_sequence
text = "It's terrific when a funny movie doesn't make smile you. What a pity!! This film is very boring and so long. It's simply painfull. The story is staggering without goal and no fun.<br /><br />You feel better when it's finished."
result = text_to_word_sequence(text)
print(result)
```

```python
from keras.preprocessing.text import one_hot
from keras.preprocessing.text import text_to_word_sequence

Text = 'It\'s terrific when a funny movie doesn\'t make smile you. What a pity!! This film is very boring and so long. It\'s simply painfull. The story is staggering without goal and no fun.<br /><br />You feel better when it\'s finished.'

Words = set(text_to_word_sequence(Text))
vocab_size = len(Words)
print(vocab_size)

result = one_hot(Text , round(vocab_size * 1.3))
print(result)
```

```python
from keras.preprocessing.text import Tokenizer
docs = corpus
tokenizer = Tokenizer()
tokenizer.fit_on_texts(docs)

print("- Word counts:")
print(tokenizer.word_counts,'\n')

print("- Document counts:")
print(tokenizer.document_count,'\n')

print("- Word index: ")
print(tokenizer.word_index,'\n')

print("- Word docs: ")
print(tokenizer.word_docs,'\n')


encoded_docsB = tokenizer.texts_to_matrix(docs, mode='binary')
encoded_docsC = tokenizer.texts_to_matrix(docs, mode='count')
encoded_docsT = tokenizer.texts_to_matrix(docs, mode='tfidf')
encoded_docsF = tokenizer.texts_to_matrix(docs, mode='freq')

print("Binary : ")
print(encoded_docsB ,'\n')

print("Count : ")
print(encoded_docsC,'\n')

print("TFIDF : ")
print(encoded_docsT,'\n')

print("Freq : ")
print(encoded_docsF)
```
